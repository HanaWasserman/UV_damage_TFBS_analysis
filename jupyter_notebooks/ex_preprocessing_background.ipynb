```python
%matplotlib inline
import pandas as pd
pd.options.mode.chained_assignment = None
import yaml
from scripts.prep_background import prep_background_regions, prep_background_dams,proc_background_dams, make_damage_model
from scripts.repair.make_repair_dictionary import make_repair_dict
```

```python
with open("config/pipeline_config.yaml", "r") as f:
    config = yaml.safe_load(f)

genome_dir = config["folders"]["genome_dir"]
main_dir = config["folders"]["main_dir"]
cpd_seq_data_dir = config["folders"]["cpd_seq_data_dir"]

kmer = config["parameters"]["kmer_length"]

archetype_file = config["files"]["archetype_file"]
dhs_file = config["files"]["dhs_file"]
```

```python
experiments = ['WT_CSB_6J_0hr_CPD_1bp_sorted','WT_CSB_6J_6hr_CPD_1bp_sorted', 'XPC_12J_NakedDNA_S22_CPD_1bp_sorted']
```

```python
#Prepare background regions
prep_background_regions(dhs_file, f"{main_dir}/data", f"{main_dir}/results/background", genome_dir)
```

```python
for exp in experiments:
    #Intersect damage count data with background regions
    proc_background_dams(dhs_file, kmer, exp, f"{main_dir}/results/background",cpd_seq_data_dir, f"{main_dir}/data", genome_dir)

    #Collate all intersected background data
    prep_background_dams(dhs_file, exp, kmer, cpd_seq_data_dir, f"{main_dir}/results/background")

    #Calculate damage formation frequencies for modeling
    make_damage_model(dhs_file, kmer, exp, f"{main_dir}/results/background")
```

```python
#Make repair kmer dictionary for repair analyses
make_repair_dict(dhs_file, f"{main_dir}/results/background", experiments[0], experiments[1])
```
