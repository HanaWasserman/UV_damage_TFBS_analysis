```python
import pandas as pd
import yaml
from scripts.repair.simulate_TFBS_repair import sim_repair
from scripts.repair.analyze_TFBS_repair_sims import analyze_repair_sims
%matplotlib inline
```

```python
with open("config/pipeline_config.yaml", "r") as f:
    config = yaml.safe_load(f)

main_dir = config["folders"]["main_dir"]

tf_window_size = config["parameters"]["tf_window_size"]
tfbs_buffer_size = config["parameters"]["tfbs_buffer_size"]
repair_window = config["parameters"]["damage_repair_analysis_window"]

dhs_file = config["files"]["dhs_file"]
motif_cluster_file = config["files"]["motif_cluster_file"]
```

```python
exps = ['WT_CSB_6J_0hr_CPD_1bp_sorted','WT_CSB_6J_6hr_CPD_1bp_sorted']
```

```python
clusters = pd.read_csv(f'{main_dir}/data/TF_cluster_motifs.csv')
cluster_records = clusters.to_records(index=False)
```

```python
for tf_cluster, tf_len, tf in cluster_records:
    #Simulate expected aggregate CPD counts at second timepoint
    for strand in ["opp","same"]:
        for p in range(-30,30):
            #NOTE bootstrap samples=100 for demonstration. Use HPC for larger bootstrap samplings
            sim_repair(100, tf, dhs_file, p, exps[0], exps[1], strand, tf_window_size, repair_window, f"{main_dir}/results/background", f"{main_dir}/results/tf_damage_data", f"{main_dir}/results/analysis/repair")
```

```python
for tf_cluster, tf_len, tf in cluster_records:
    analyze_repair_sims(tf, tf_len, exps[0], exps[1], 100, dhs_file, f"{main_dir}/results/analysis/repair")
```
